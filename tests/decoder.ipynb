```python
def decode_px4_custom_mode(custom_mode: int):
    """
    Decode PX4 custom_mode into main mode and sub mode
    """

    main_mode = (custom_mode >> 16) & 0xFF
    sub_mode  = (custom_mode >> 24) & 0xFF

    MAIN_MODES = {
        1: "MANUAL",
        2: "ALTCTL",
        3: "POSCTL",
        4: "AUTO",
        5: "ACRO",
        6: "OFFBOARD",
        7: "STABILIZED",
        8: "RATTITUDE",
    }

    AUTO_SUBMODES = {
        1: "READY",
        2: "TAKEOFF",
        3: "LOITER",
        4: "MISSION",
        5: "RTL",
        6: "LAND",
        7: "RTGS",
        8: "FOLLOW_TARGET",
    }

    main_str = MAIN_MODES.get(main_mode, f"UNKNOWN({main_mode})")

    if main_mode == 4:  # AUTO
        sub_str = AUTO_SUBMODES.get(sub_mode, f"UNKNOWN({sub_mode})")
    else:
        sub_str = None

    return {
        "custom_mode": custom_mode,
        "hex": hex(custom_mode),
        "main_mode": main_mode,
        "main_mode_str": main_str,
        "sub_mode": sub_mode,
        "sub_mode_str": sub_str,
    }
```

```python
def encode_px4_mode_from_names(main_mode_str: str, sub_mode_str: str = None):
    """
    Input: mode names → Output: custom_mode int + hex
    """

    MAIN_MODES = {
        "MANUAL": 1,
        "ALTCTL": 2,
        "POSCTL": 3,
        "AUTO": 4,
        "ACRO": 5,
        "OFFBOARD": 6,
        "STABILIZED": 7,
        "RATTITUDE": 8,
    }

    AUTO_SUBMODES = {
        "READY": 1,
        "TAKEOFF": 2,
        "LOITER": 3,
        "MISSION": 4,
        "RTL": 5,
        "LAND": 6,
        "RTGS": 7,
        "FOLLOW_TARGET": 8,
    }

    main_mode_str = main_mode_str.upper()
    main_mode = MAIN_MODES.get(main_mode_str)

    if main_mode is None:
        raise ValueError(f"Invalid main mode: {main_mode_str}")

    sub_mode = 0

    if main_mode_str == "AUTO":
        if sub_mode_str is None:
            raise ValueError("AUTO mode requires sub_mode")
        sub_mode_str = sub_mode_str.upper()
        sub_mode = AUTO_SUBMODES.get(sub_mode_str)

        if sub_mode is None:
            raise ValueError(f"Invalid AUTO sub mode: {sub_mode_str}")

    custom_mode = (sub_mode << 24) | (main_mode << 16)

    return {
        "input": (main_mode_str, sub_mode_str),
        "custom_mode": custom_mode,
        "hex": hex(custom_mode)
    }
```

```python
encode_px4_mode_from_names("AUTO", "READY")
```
